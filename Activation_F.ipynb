```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
x = np.linspace(-10,10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
  x = np.linspace(-1,+1)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show()
```

```python
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)

  return x1
  x = np.linspace(-10,10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :RELU')
plt.show()
```
